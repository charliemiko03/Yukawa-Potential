```python
from scipy.constants import m_p, m_n
mP = m_p  # kg
mP = mP*1e30  # change order scale
mP = mP/1.783  # MeV/c^2, but c = 1 in natural units
print(f'mP = {mP} MeV')  # MeV

mN = m_n  # kg
mN = mN*1e30  # change order scale
mN = mN/1.783  # MeV/c^2, but c = 1 in natural units
print(f'mN = {mN} MeV')  # MeV
```

```python
def V(r):
    return np.where(r <= a, -V0, np.exp(-lamda*r)/r)
```

```python
def centrifugal(r):
    return hbar**2*L*(L+1)/(2*m*r**2)
```

```python
def V_eff(r):
    return centrifugal(r) + V(r)
```

```python
hbar, m, a = 1.0, 1.0, 1.0
V0 = 10.0
lamda = 0.2
r = np.linspace(0.1, 25, 1000)
```

```python
L = 1

plt.figure()
plt.plot(r, centrifugal(r), 'b-', label='Centrifugal')
plt.axhline(0, color='k', linestyle='--')
plt.ylim(-0.1, 1.5)
plt.xlabel('$r$ (fm)')
plt.ylabel('$V_{eff}(r)$ (MeV)')
plt.legend()
plt.show()
```

```python
L = 1

plt.figure()
plt.plot(r, V(r), 'g-', label='Potential')
plt.axhline(0, color='k', linestyle='--')
plt.xlim(0, 5)
plt.ylim(-V0-0.5, 2.0)
plt.xlabel('$r$ (fm)')
plt.ylabel('$V(r)$ (MeV)')
plt.legend()
plt.show()
```

```python
L = 1

plt.figure()
plt.plot(r, V_eff(r), 'b-', label='$V_{eff}$')
plt.axhline(0, color='k', linestyle='--')
#plt.axhline(-V0, color='r', linestyle='--', label='$-V_0$')
plt.xlim(0, 5)
plt.ylim(-V0-0.5, 5.0)
plt.xlabel('$r$ (fm)')
plt.ylabel('$V_{eff}(r)$ (MeV)')
plt.legend()
plt.show()
```
